```python
import numpy as np
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
# 선형회귀를 이용한 모델 구성
x_train = np.array([1, 2, 3])
y_train = np.array([1, 2, 3])

model = Sequential() # 빈 모델객체 생성
model.add(Dense(1, input_dim=1)) # 1차원, 가중치도 1

sgd = SGD(lr=0.1) # 경사를 지정
# loss - 비용함수
# MSE를 이용한 cost func 를 구할거고,
# optimizer는 경사하강법 알고리즘중 하나인 SGD 알고리즘을 쓰겠다는거임
model.compile(loss='mse', optimizer=sgd)

# prints summary of the model to the terminal
model.summary()

model.fit(x_train, y_train, epochs=200)

y_predict = model.predict(np.array([5]))
print(y_predict)

# 같은 값을 주고 실행을 해도 loss 값은 전부 다르게 랜덤으로 출력된다.
```

```python
# 선형회귀를 이용한 모델 구성

# x_train = np.array([73, 93, 89, 96, 73],
#                     [80, 88, 91, 98, 66],
#                     [75, 93, 90, 100, 70])
# 여섯번째 학생의 항목을 3개를 입력하기 때문에 3x5 행렬로 작성한다.
# 아니면 x_train.T 로 해도 되긴 됨

x_train = np.array([[73, 80, 75],
                   [93, 88, 93],
                   [89, 91, 90],
                   [96, 98, 100],
                   [73, 66, 70]])
y_train = np.array([152, 185, 180, 196, 142])

model = Sequential() # 빈 모델객체 생성
model.add(Dense(1, input_dim=3)) # 1차원, 가중치도 1

sgd = SGD(lr=0.0000001) # 경사를 지정
# 여기서 경사가 0.1이면 너무 값이 커서 nan으로 뜨기 때문에
# loss 를 굉장히 낮게 설정한다.
model.compile(loss='mse', optimizer=sgd)

# prints summary of the model to the terminal
model.summary()

model.fit(x_train, y_train, epochs=200)

sixth_student = np.array([60, 99, 57])
y_predict = model.predict(np.expand_dims(sixth_student, axis=0)) # axis=0 은 가장 앞 차원을 뜻함

print("-" * 50)
print(y_predict)

# 같은 값을 주고 실행을 해도 loss 값은 전부 다르게 랜덤으로 출력된다.
```

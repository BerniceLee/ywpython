```python
# 이 가상환경에는 konlpy 가 없어서 명령어로 바로 설치해준다.

!pip install konlpy
from konlpy.tag import Okt
import codecs
```

```python
# readFp = codecs.open("./kowiki-latest-pages-articles/wiki.wp2txt", "r")
readFp = codecs.open("./kowiki-latest-pages-articles/wiki.wp2txt", "r", encoding="utf-16")
gubun_file = "wifi.gubun"

# writeFp = open(gubun_file, "w")
writeFp = open(gubun_file, "w", encoding="utf-16")
okt = Okt()
```

```python
i=0

while True :
    line = readFp.readline()
    if not line : 
        break    
    if i % 20000 == 0 :
        print("current : " + str(i))
        
    i+=1
    
    # okt의 pos 메소드 : 품사단위로 텍스트 추출
    malist = okt.pos(line, norm=True, stem=True)
    r = []
    
    for word in malist :
        # 두번째 인덱스에 있는 것들이 조사, 어미, 구분자가 아닌 경우 공백 추가
        if not word[1] in ["Josa", "Eomi", "Punctuation"] :
            writeFp.write(word[0] + " ")
            
writeFp.close()            
```

여기서 에러 주의해야하는게, Text8Corpus 가 utf-8만 되도록 만들어져있음.

그래서 위에서 utf-16 으로 풀어주게 되면, 여기서부터 에러가 나는것임

그래서 위에서 풀어준 파일을 utf-8로 변환 후 진행해야한다.

```python
# 입력 스트림과 출력 스트림을 연다
input = open("wifi.gubun", "rt", encoding="utf-16")
output = open("wiki.gubun", "wt", encoding="utf-8")

# 유니코드 데이터 조각들을 스트리밍한다
with input, output:
    while True:
        # 데이터 조각을 읽고
        chunk = input.read(4096)
        if not chunk:
            break
        # 수직 탭을 삭제한다
        chunk = chunk.replace("\u000B", "")
        # 데이터 조각을 쓴다
        output.write(chunk)
```

utf-8 로 변환한 "wiki.gubun" 파일을 이제 text8corpus 로 풀어놓고 model 로 저장한다.

```python
from gensim.models import word2vec

sentences = word2vec.Text8Corpus("wiki.gubun")
model = word2vec.Word2Vec(sentences, sg=1, size=100, window=5)
model.save("./kowiki-latest-pages-articles/wiki.model")

'''
런타임에러 신경안써도됨
보카학습 첨부터 1980만까지 시키기 싫어가지고 그냥 교수님꺼 모델 받아서 진행함
'''
```

이제 모델이 잘 만들어지고 이상 없는게 확인이 되었으니까, 이거로 문장을 실제로 찍어보자.

```python
model = word2vec.Word2Vec.load("./kowiki-latest-pages-articles/wiki.model")
okt = Okt()

def print_emergency(text) : 
    print(text)
    # 전달된 문장의 형태소 분석하기
    node = okt.pos(text, norm=True, stem=True)
    
    for word, form in node :
        # 필요한 형태소만 추출하기
        if form == "Noun" or form == "Verb" or form == "Adjective" or form == "Adverb" :
            # 급하다와 비슷한 단어
            print("-", word, ":", model.wv.similarity(word, "급하다"))
            
print_emergency("컴퓨터에 문제가 생겼어요. 빨리 해결해야 하는 문제가 있어서 지금 확인 해주셔야해요.")
print_emergency("사용 방법을 잘 모르겠어요")
print_emergency("술이 너무 응급합니다 살려주세요 만취로로 절 데려다주세요")
```

국립국어원 언어정보나눔터에서 말뭉치 파일을 아무거나 가져옴

나는 주제대화_PC방문화, 전자전사자료 로 가져옴

이 파일을 까보면 html 구조로 되어있는데,

이 파일을 근거로 PC방에간 남자들의 문체를 학습시켜보자.

```python
import codecs
from bs4 import BeautifulSoup
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.layers import LSTM
from keras.optimizers import RMSprop
from keras.utils.data_utils import get_file
import numpy as np
import random, sys

fp = codecs.open("./bs4_pcroomchat/BEXX0004.txt", "r", encoding="utf-16")
soup = BeautifulSoup(fp, "html.parser")
body = soup.select_one("body")
text = body.getText() + " " 
text[3]
print("코퍼스의 길이 : ", len(text))

# 문자를 하나하나 읽어 들이고 ID 붙이기
chars = sorted(list(set(text)))
print("사용되고 있는 문자의 수 : ", len(chars))
char_indices = dict((c, i) for i,c in enumerate(chars)) # 문자 -> ID
indices_char = dict((i, c) for i,c in enumerate(chars)) # ID -> 문자

# 텍스트를 maxlen 개의 문자로 자르고, 다음에 오는 문자 등록하기
maxlen = 20
step = 3 
sentences = []
next_chars = []

for i in range(0, len(text) - maxlen, step) :
    sentences.append(text[i: i + maxlen])
    next_chars.append(text[i + maxlen])

print("학습할 구문의 수 : ", len(sentences))
print("텍스트를 ID 벡터로 변환합니다......")

X = np.zeros((len(sentences), maxlen, len(chars)), dtype=np.bool)
y = np.zeros((len(sentences), len(chars)), dtype=np.bool)

for i, sentence in enumerate(sentences) :
    for t, char in enumerate(sentence) :
        X[i, t, char_indices[char]] = 1
    y[i, char_indices[next_chars[i]]] = 1

print("ID 벡터 변환 완료")  

# LSTM 모델 구축
print("모델을 구축합니다.....")
model = Sequential()
model.add(LSTM(128, input_shape = (maxlen, len(chars))))
model.add(Dense(len(chars)))
model.add(Activation('softmax'))
optimizer = RMSprop(lr = 0.01)
model.compile(loss='categorical_crossentropy', optimizer = optimizer)

# 후보를 배열에서 꺼내기
def sample(preds, temperature = 1.0) :
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    probas = np.random.multinomial(1, preds, 1)
    return np.argmax(probas)

# 학습시키고 텍스트 생성하기 반복
for iteration in range(1, 60):
    print()
    print('-' * 50)
    print('반복 =', iteration)
    model.fit(X, y, batch_size=128, nb_epoch=1)
    
    # 임의의 시작 텍스트 선택하기
    start_index = random.randint(0, len(text) - maxlen - 1)
    
    # 다양한 다양성의 문장 생성
    for diversity in [0.2, 0.5, 1.0, 1.2]:
        print()
        print('--- 다양성 = ', diversity)
        generated = ''
        sentence = text[start_index: start_index + maxlen]
        generated += sentence
        print('--- 시드 = ' + sentence + '"')
        sys.stdout.write(generated)
        
        # 시드를 기반으로 텍스트 자동 생성
        for i in range(400):
            x = np.zeros((1, maxlen, len(chars)))
            for t, char in enumerate(sentence):
                x[0, t, char_indices[char]] = 1.
                
            # 다음에 올 문자를 예측하기
            preds = model.predict(x, verbose=0)[0]
            next_index = sample(preds, diversity)
            next_char = indices_char[next_index]
            
            # 출력하기
            generated += next_char
            sentence = sentence[1:] + next_char
            sys.stdout.write(next_char)
            sys.stdout.flush()
            
        print()
```

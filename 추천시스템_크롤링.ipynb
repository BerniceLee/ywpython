```python
%%time
dic={}

for j in range(1,4950):
    url='http://www.enuri.com/enuripc/requestEstimateDetail.jsp?idx={}&page=1&buy_use=&wish_price='.format(5000-int(j))
    driver.get(url)
    q={}
    price_=[]
    for i in range(1,6):
        try:
            category_url_2='#compProdDetailTbody>tr:nth-child({})>th:nth-child(1)'.format(i)
            goods_url_2='#compProdDetailTbody > tr:nth-child({}) > td:nth-child(2) > p.detxt'.format(i)
            category_url_5='#compProdDetailTbody>tr:nth-child({})>th:nth-child(4)'.format(i)
            goods_url_5='#compProdDetailTbody > tr:nth-child({}) > td:nth-child(5) > p.detxt'.format(i)
            purpose_url='#reqEstView > tbody > tr:nth-child(3) > td:nth-child(2)'
            
            price_url_2='#compProdDetailTbody > tr:nth-child({}) > td:nth-child(2) > p.lowprice '.format(i)
            price_url_5='#compProdDetailTbody > tr:nth-child({}) > td:nth-child(5) > p.lowprice '.format(i)
            
            price_2=driver.find_element_by_css_selector(price_url_2).text
            price_5=driver.find_element_by_css_selector(price_url_5).text
                     
            cate_2=driver.find_element_by_css_selector(category_url_2).text
            goods_2=driver.find_element_by_css_selector(goods_url_2).text
            
            cate_5=driver.find_element_by_css_selector(category_url_5).text
            goods_5=driver.find_element_by_css_selector(goods_url_5).text
      
            purpose=driver.find_element_by_css_selector(purpose_url).text
            

            if cate_2 in ['CPU', '메인보드', 'SSD', 'VGA' , '메모리' , '파워' , 'HDD']:
                print(i,cate_2,price_2)
                price_.append(int(price_2.replace('원','').replace(',','')))
            if cate_2 in ['CPU', '메인보드', 'SSD', 'VGA' , '메모리' , '파워' , 'HDD']:
                print(i,cate_5,price_5)
                price_.append(int(price_5.replace('원','').replace(',','')))    
              
    
            q[cate_2]=(goods_2)
            q[cate_5]=(goods_5)
            
            
        except:
            print('★★★★★★★★★★★★★★★ NO ELEMENTS IN ROW ★★★★★★★★★★★★★★★')
    
    q['price']=sum(price_)
    q['purpose']=purpose
    dic['user'+str(j)]=q
    
    # 10000 - 5000 : Wall time: 1h 20min 31s
    # 5000 - 50 : Wall time: 1h 18min 38s
    
```
